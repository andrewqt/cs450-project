```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('out2.txt')
```

```python
df
```

```python
pd.reset_option('display.max_rows')
grouped_df = df.groupby(['test', 'algo', 'nodes'])['total time'].mean()
```

```python
for time in df.columns[4:]:
    grouped_df = df.groupby(['test', 'algo', 'nodes'])[time].mean()
    fig, ax = plt.subplots()

    line_data = grouped_df['line_tree']

    for algo, group in line_data.groupby('algo'):
        print(group[algo].index)
        ax.plot(group[algo].index, group[algo].values, label=algo)

    # Customize the plot
    ax.set_xlabel('nodes')
    ax.set_ylabel(time)
    ax.set_title(f'{time} for line tree  ')
    ax.legend()

    # Show the plot
    # plt.xscale('log', base=2)
    # plt.yscale('log', base=2)
    plt.show()
```
